```python
# Se Modifica el formato de los datos (Coerción de datos, quitar comas)

forbes = forbes.replace(',','',regex= True) # regex: expresión regular
forbes.head()
```

```python
# Eliminar signo pesos 
forbes = forbes.replace('\$','',regex= True) # regex: expresión regular
forbes.head()
```

```python
# Remplazamos el '-' por NaN que es el missing de Pandas

forbes['Assets'] = forbes['Assets'].replace('-',float('nan')) 
```

```python
# la funcion groupby permite agrupar datos en función de los demás. Es decir, hacer el análisis del DataFrame en función de una de las columnas.
forbes.groupby('Country').size()
#La función Size se utiliza para definir la forma de una matriz, es decir, las dimensiones de la matriz: número de filas y número de columnas
```

```python
# La función pandas fillna () es útil para completar los valores faltantes en las columnas de un DataFrame de pandas.
# La Funcion lambda se definen como una línea que ejecuta una sola expresión.
# La función mean() devuelve la media aritmética de los datos de una muestra o población.
# inplace=True realiza la operación de eliminado sobre el mismo DataFrame, en lugar de crear y devolver uno nuevo.

forbes["Assets"].fillna(forbes.groupby('Country')["Assets"].transform(lambda x : x.mean()), inplace=True)
forbes[forbes['Country'] == 'Greece']
```
