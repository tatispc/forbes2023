```python
# la funcion groupby permite agrupar datos en función de los demás. Es decir, hacer el análisis del DataFrame en función de una de las columnas.
forbes23.groupby('Country').size()
#La función Size se utiliza para definir la forma de una matriz, es decir, las dimensiones de la matriz: número de filas y número de columnas
```

```python
# La función pandas fillna () es útil para completar los valores faltantes en las columnas de un DataFrame de pandas.
# La Funcion lambda se definen como una línea que ejecuta una sola expresión.
# La función mean() devuelve la media aritmética de los datos de una muestra o población.
# inplace=True realiza la operación de eliminado sobre el mismo DataFrame, en lugar de crear y devolver uno nuevo.

forbes23["Assets"].fillna(forbes23.groupby('Country')["Assets"].transform(lambda x : x.mean()), inplace=True)
forbes23[forbes23['Country'] == 'Greece']
```

```python
#Percentiles

print(np.percentile(forbes23['Profits'],[25,50,75]))

print(forbes23['Profits'].quantile([0.25,0.5,0.75]))
```

```python
# Valorar el 1% de las empresas con mas ganancias en el mundo

p99p = np.percentile(forbes23['Profits'],99)
print(p99p)
```

```python
# Filtrar de mi DF usando la condición de 'Profits' >  p99p

forbes23[forbes23['Profits'] > p99p][['Company','Country','Profits']] 
```

```python
# 1% de las empresas con mayor valor del mercado

p99mv = forbes23['Market_value'].quantile(0.99)
print(p99mv)
```

```python
forbes23[forbes23['Market_value']>p99mv][['Company','Country','Market_value']] 
```

```python
# Del 1% mas alto en valor de mercado, cuáles están en el 1% con mas ganancias
# Operadores lógicos: y &, o |

forbes23[(forbes23['Market_value']>p99mv) & (forbes23['Profits']>p99p)][['Company','Country','Market_value','Profits']] 
```

```python
def cv(data):
    media = np.mean(data)
    sd = np.std(data)
    coef = (sd/media)*100
    return coef
```

```python
cv(forbes23['Sales']) 
```

```python
forbes23.iloc[:,3:].apply(func=cv,axis = 0)
```

```python
#pip install matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar plot grafico de barras
table_suda=f_suda['Country'].value_counts()
table_suda

plt.bar(table_suda.index,table_suda.values, color = 'purple', alpha = 0.9, edgecolor = 'white') #Crea el gráfico de barras
plt.title('Frecuencia de empresas sudamericanas')
plt.xlabel('Paises sudamericanos')
plt.ylabel('Frecuencia absoluta')
plt.grid(color = 'gray',alpha = 0.3, linestyle = 'solid')

plt.show() #Imprime el gráfico
```

```python
table_europe = f_europe['Country'].value_counts()


plt.bar(table_europe.index,table_europe.values, color = 'orange', alpha = .9, edgecolor = 'white') #Crea el gráfico de barras
plt.title('Frecuencia de empresas sudamericanas')
plt.xlabel('Paises sudamericanos')
plt.ylabel('Frecuencia absoluta')
plt.grid(color = 'gray',alpha = 0.3, linestyle = 'solid')

plt.show() #Imprime el gráfico
```

```python
# pie chart

plt.pie(table_suda.values,labels=table_suda.index)
plt.title('Segmento de empresas por paises')

plt.show()
```
